We can see there is a great number of datas in this dataset. The table of the data and its meaning is avaliable on the GitHub page.

The data is provided in three levels: world data, continent data, and country data. We are first going to divide them into these three parts. (Actually one extra, including ER and the Ship ***Princess***)

```python
world_data = pd.DataFrame(columns = data.columns)
continent_data = pd.DataFrame(columns = data.columns)
country_data = pd.DataFrame(columns = data.columns)
others_data = pd.DataFrame(columns = data.columns)
```

```python
world_iso_code = ['OWID_WRL']
continent_iso_code = ['OWID_AFR', 'OWID_ASI', 'OWID_EUR', 'OWID_NAM', 'OWID_SAM', 'OWID_OCE']
others_iso_code = ['OWID_INT', 'OWID_EUN']
```

```python
iso_code = list(dict.fromkeys(data["iso_code"].tolist()))
iso_code
```

```python
country_iso_code = [i for i in iso_code if i not in world_iso_code
                                       and i not in continent_iso_code
                                       and i not in others_iso_code]
country_iso_code
```

```python
world_data = data.loc[data['iso_code'].isin(world_iso_code)]
world_data = world_data.reset_index()
world_data = world_data.drop(columns = ['index'])
world_data
```

```python
continent_data = data.loc[data['iso_code'].isin(continent_iso_code)]
continent_data = continent_data.reset_index()
continent_data = continent_data.drop(columns = ['index'])
continent_data
```

```python
country_data = data.loc[data['iso_code'].isin(country_iso_code)]
country_data = country_data.reset_index()
country_data = country_data.drop(columns = ['index'])
country_data
```

```python
others_data = data.loc[data['iso_code'].isin(others_iso_code)]
others_data = others_data.reset_index()
others_data = others_data.drop(columns = ['index'])
others_data
```

```python
world_data.to_csv('world_data.csv')
continent_data.to_csv('continent_data.csv')
country_data.to_csv('country_data.csv')
others_data.to_csv('others_data.csv')
```

```python
world_data_dict = {}
continent_data_dict = {}
country_data_dict = {}
eun_data_dict = {}
international_data_dict = {}
```

```python
for iso_code in world_iso_code:
    temp_df_columns = world_data.columns.tolist()
    temp_df_columns.remove('date')
    temp_df = pd.DataFrame(columns = temp_df_columns)
    
    temp_now_loc = world_data.loc[world_data['iso_code'] == iso_code]
    temp_now_loc = temp_now_loc.drop(columns = ['iso_code'])
    temp_now_loc = temp_now_loc.set_index('date')
    
    world_data_dict[iso_code] = temp_now_loc
    
    temp_now_loc.to_csv('./world_data/' + iso_code + '.csv')
    
print(world_data_dict)
```

```python
for iso_code in continent_iso_code:
    temp_df_columns = continent_data.columns.tolist()
    temp_df_columns.remove('date')
    temp_df = pd.DataFrame(columns = temp_df_columns)
    
    temp_now_loc = continent_data.loc[continent_data['iso_code'] == iso_code]
    temp_now_loc = temp_now_loc.drop(columns = ['iso_code'])
    temp_now_loc = temp_now_loc.set_index('date')
    
    continent_data_dict[iso_code] = temp_now_loc
    
    temp_now_loc.to_csv('./continent_data/' + iso_code + '.csv')
    
print(continent_data_dict)
```

```python
for iso_code in country_iso_code:
    temp_df_columns = country_data.columns.tolist()
    temp_df_columns.remove('date')
    temp_df = pd.DataFrame(columns = temp_df_columns)
    
    temp_now_loc = country_data.loc[country_data['iso_code'] == iso_code]
    temp_now_loc = temp_now_loc.drop(columns = ['iso_code'])
    temp_now_loc = temp_now_loc.set_index('date')
    
    country_data_dict[iso_code] = temp_now_loc
    
    temp_now_loc.to_csv('./country_data/' + iso_code + '.csv')
    
print(country_data_dict)
```

```python
temp_df_columns = others_data.columns.tolist()
temp_df_columns.remove('date')
temp_df = pd.DataFrame(columns = temp_df_columns)

eun = others_data.loc[others_data['iso_code'] == 'OWID_EUN']
eun = eun.drop(columns = ['iso_code'])
eun = eun.set_index('date')

eun_data_dict['OWID_EUN'] = eun

eun.to_csv('./others_data/' + 'OWID_EUN' + '.csv')

print(eun_data_dict)

international = others_data.loc[others_data['iso_code'] == 'OWID_INT']
international = international.drop(columns = ['iso_code'])
international = international.set_index('date')

international_data_dict['OWID_INT'] = international

international.to_csv('./others_data/' + 'OWID_INT' + '.csv')
```
